18. จงเขียนโปรแกรมสุ่มเลขจำนวนเต็มในช่วง 10-100 ออกมาอยู่ใน matrix 5x5 และใช้ If-else เช็คทีละตัวว่าตัวใดบ้างที่มีค่ามากกว่า 50 และปริ้นออกมา

```python
import numpy as np
xlst = list()
x = np.random.randint(10,101,(5,5))
for i in range(5):
    for j in range(5):
        if (x[i][j] > 50):
            xlst.append(x[i,j])
print(f'Matrix 5x5\n{x}\n\nValue > 50\n{np.array(xlst)}')
```

19. จงเขียนโปรแกรมสุ่มเลขจำนวนเต็มในช่วง 10-100 ออกมาอยู่ใน matrix 5x5 และใช้ conditional selection เช็คว่าตัวใดบ้างมีค่ามากกว่า 40 นำมาใส่ List ที่สร้างใหม่ และปริ้นออกมา

```python
import numpy as np
y = np.random.randint(10,101,(5,5))
lst = y[y>40].tolist()
print(f'Matrix 5x5\n{y}\n\nValue > 40\n{lst}')
```

20. เขียนโปรแกรมรับค่าจากผู้ใช้ให้ป้อนเลข 9 ตัว แต่ละตัวต้องไม่ซ้ำกันและมีค่าอยู่ในช่วง 10-100 นำเลข 9 ตัวนั้นมาทำเป็น matrix 3x3 และ transport และหาเลขอยู่ตำแหน่งที่ (2,2)

```python
import numpy as np
num = list()
for i in range(9):
    while True:
        temp = int(input(f'กรุณาใส่เลขตัวที่ {i+1} ที่มีค่า 10-100 และไม่ซ้ำกัน: '))
        checkrep = True
        for j in num:
            if (temp == j):
                checkrep = False
        if (temp >= 10 and temp <= 100 and checkrep == True):
            num.append(temp)
            break
print(f'''
Matrix 3x3
{np.array(num).reshape(3,3)}

Transport
{np.array(num).reshape(3,3).T}

Array (2,2)
{np.array(num).reshape(3,3).T[2,2]}''')
```
